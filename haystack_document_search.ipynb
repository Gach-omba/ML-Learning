```python
dataset=pd.read_csv('/content/S08_question_answer_pairs.txt',sep='\t')
```

```python
# select the article title and question
df=dataset[["ArticleTitle","Question"]]
df.head()
```

```python
df.rename(columns={"Question":"content"},inplace=True)
```

```python
# check column details to see if there are nulls
df.head()
```

```python
# delte the null columns
df.dropna(axis=0, inplace=True)
```

```python
df.info()
```

```python
## in each project we need to have to have two things, the document store object and the document retriever
# we are using postgress as backend so we use faiss
```

```python
# create the document store
# always set return embeddings as false to reduce the size of the dataset
from haystack.document_stores import FAISSDocumentStore
# create the retriever
from haystack.nodes import EmbeddingRetriever
document_store = FAISSDocumentStore(
    sql_url="sqlite:///C:\\Users\\User Pc\\Projects\\traindb",  # Correct format
    faiss_index_factory_str="Flat",
    return_embedding=False
)
```

```python
# creating a retriever
retriever_faiss= EmbeddingRetriever(document_store=document_store,
                                    embedding_model='distilroberta-base-msmarco-v2',
                                    model_format='sentence_transformers', use_gpu=False)

```

```python
# writing to the database
document_store.write_documents(df.to_dict(orient='records'))
```

```python
# update the embeddings
document_store.update_embeddings(retriever=retriever_faiss)
```

```python
# try to retrieve
df.head()
```

```python
retriever_faiss.retrieve("Did his mother die of pneumonia?", top_k=15)
```

```python
# adding new documents to the database
dict_sample={'ArticleTitle':["My Dream My Life"], "content":"Life full of love"}
```

```python
df_new=pd.DataFrame(dict_sample)
df_new
```

```python
document_store.write_documents(df_new.to_dict(orient='records')) # inserting a new record
```

```python
# if you run :
retriever_faiss.retrieve("Life full of love", top_k=15)
# it won't return the exact results since we haven't updated the embeddings
```
